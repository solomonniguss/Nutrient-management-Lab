```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np

# Load your Excel data into a DataFrame
data = pd.read_excel('Growers 2023 24.xlsx', sheet_name='2022')

# Calculate the mean and standard error for each Prate and IRR combination
mean_data = data.groupby(['Prate', 'IRR'], as_index=False).agg({'Yield': ['mean', 'sem']})
mean_data.columns = ['Prate', 'IRR', 'Yield', 'Yield_SE']

# Define the colors for the bars
colors = ['#1f77b4', '#ff7f0e']  # Colors for the different irrigation levels

# Create a scatter plot with regression lines and error bars
plt.figure(figsize=(10, 8))

# Scatter plot for mean Yield values with error bars
for irr in mean_data['IRR'].unique():
    subset = mean_data[mean_data['IRR'] == irr]
    plt.errorbar(subset['Prate'], subset['Yield'], yerr=subset['Yield_SE'], fmt='o', 
                 label=f'{irr} Irrigation', capsize=5, color=colors.pop(0))

# Adding regression lines
irrigation_levels = mean_data['IRR'].unique()
for irr in irrigation_levels:
    subset = mean_data[mean_data['IRR'] == irr]
    X = subset['Prate'].values.reshape(-1, 1)
    y = subset['Yield'].values
    model = LinearRegression().fit(X, y)
    y_pred = model.predict(X)
    plt.plot(subset['Prate'], y_pred, label=f'{irr} Irrigation Fit (R²={r2_score(y, y_pred):.2f})')

plt.title('Mean Yield vs P rate for Different Irrigation Levels', fontsize=15)
plt.xlabel('P rate (Ibs/acre)', fontsize=15)
plt.ylabel('Yield (Ibs/acre)', fontsize=15)
plt.xticks(rotation=0, fontsize=15)
plt.yticks(fontsize=15)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['left'].set_color('black')
plt.legend(title='Irrigation Level', fontsize=12)
plt.ylim(2000, 10000)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
import numpy as np

# Load your Excel data into a DataFrame
data = pd.read_excel('Growers 2023 24.xlsx', sheet_name='2022')

# Calculate the mean and standard error for each Prate and IRR combination
mean_data = data.groupby(['Prate', 'IRR'], as_index=False).agg({'Yield': ['mean', 'sem']})
mean_data.columns = ['Prate', 'IRR', 'Yield', 'Yield_SE']

# Define the colors for the bars
colors = ['#1f77b4', '#ff7f0e']  # Colors for the different irrigation levels

# Create a scatter plot with regression lines and error bars
plt.figure(figsize=(10, 8))

# Scatter plot for mean Yield values with error bars
for irr in mean_data['IRR'].unique():
    subset = mean_data[mean_data['IRR'] == irr]
    plt.errorbar(subset['Prate'], subset['Yield'], yerr=subset['Yield_SE'], fmt='o', 
                 label=f'{irr} Irrigation', capsize=5, color=colors.pop(0))

# Function to fit and plot models
def fit_and_plot_model(X, y, degree, label):
    poly = PolynomialFeatures(degree)
    X_poly = poly.fit_transform(X)
    model = LinearRegression().fit(X_poly, y)
    y_pred = model.predict(X_poly)
    r2 = r2_score(y, y_pred)
    plt.plot(X, y_pred, label=f'{label} (R²={r2:.2f})')
    return r2

# Adding quadratic regression lines
irrigation_levels = mean_data['IRR'].unique()
colors = ['#1f77b4', '#ff7f0e']  # Reset colors for lines
for irr in irrigation_levels:
    subset = mean_data[mean_data['IRR'] == irr]
    X = subset['Prate'].values.reshape(-1, 1)
    y = subset['Yield'].values
    
    # Quadratic model
    fit_and_plot_model(X, y, 2, f'{irr} Quadratic Fit')

plt.title('Mean Yield vs P rate for Different Irrigation Levels', fontsize=15)
plt.xlabel('P rate (Ibs/acre)', fontsize=15)
plt.ylabel('Yield (Ibs/acre)', fontsize=15)
plt.xticks(rotation=0, fontsize=15)
plt.yticks(fontsize=15)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['left'].set_color('black')
plt.legend(title='Irrigation Level', fontsize=12)
plt.ylim(2000, 10000)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
import scipy.stats

# Load your Excel data into a DataFrame
# Replace 'your_excel_file.xlsx' with the actual path to your Excel file
data = pd.read_excel('Growers 2023 24.xlsx',sheet_name='1st and 2nd')


# Calculate the mean and standard error for each Prate and IRR combination
mean_data = data.groupby(['Prate', 'IRR'], as_index=False).agg({'2nd_yield Ibs/acr': ['mean', 'sem']})
mean_data.columns = ['Prate', 'IRR', '2nd_yield Ibs/acr', '2nd_yield Ibs/acr_SE']

# Define the colors for the bars
colors = ['#1f77b4', '#ff7f0e']  # Colors for the different irrigation levels

# Create a scatter plot with error bars
plt.figure(figsize=(10, 8))

# Scatter plot for mean 2nd_yield Ibs/acr values with error bars
for irr in mean_data['IRR'].unique():
    subset = mean_data[mean_data['IRR'] == irr]
    plt.errorbar(subset['Prate'], subset['2nd_yield Ibs/acr'], yerr=subset['2nd_yield Ibs/acr_SE'], fmt='o', 
                 label=f'{irr} Irrigation', capsize=5, color=colors.pop(0))

# Function to fit a piecewise linear model
def piecewise_linear(x, x0, y0, k1, k2):
    return np.piecewise(x, [x < x0], [lambda x: k1 * x + y0 - k1 * x0, lambda x: k2 * x + y0 - k2 * x0])

# Function to calculate the confidence interval
def conf_interval(x, func, popt, pcov):
    alpha = 0.05  # 95% confidence interval
    n = len(x)  # Number of data points
    p = len(popt)  # Number of parameters
    dof = max(0, n - p)  # Degrees of freedom
    tval = np.abs(scipy.stats.t.ppf(alpha / 2, dof))  # Student's t value for the dof

    # Calculate the confidence interval
    y_hat = func(x, *popt)
    jacobian = np.array([scipy.optimize.approx_fprime(popt, lambda p: func(xi, *p), epsilon=1e-8) for xi in x])
    ci = tval * np.sqrt(np.diag(jacobian @ pcov @ jacobian.T))

    return y_hat, y_hat - ci, y_hat + ci

# Adding piecewise linear models
irrigation_levels = mean_data['IRR'].unique()
colors = ['#1f77b4', '#ff7f0e']  # Reset colors for lines
for irr in irrigation_levels:
    subset = mean_data[mean_data['IRR'] == irr]
    X = subset['Prate'].values
    y = subset['2nd_yield Ibs/acr'].values

    # Initial guess for parameters
    p0 = [np.median(X), np.median(y), 1, 1]

    # Fit piecewise linear model
    params, pcov = curve_fit(piecewise_linear, X, y, p0)
    y_hat = piecewise_linear(X, *params)

    # Calculate R-squared
    residuals = y - y_hat
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((y - np.mean(y))**2)
    r2 = 1 - (ss_res / ss_tot)

    # Calculate confidence interval
    y_hat, ci_lower, ci_upper = conf_interval(X, piecewise_linear, params, pcov)

    # Plot the piecewise linear fit and confidence interval
    plt.plot(X, y_hat, label=f'{irr} Plateau Fit (R²={r2:.2f})', color=colors.pop(0))
    plt.fill_between(X, ci_lower, ci_upper, color='grey', alpha=0.2)

    # Annotate the formula on the plot
    plt.text(0.5, 0.9, 'y = k1x + y0 - k1x0 (if x < x0)\ny = k2x + y0 - k2x0 (if x ≥ x0)',
             horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=12)

plt.title('Mean 2nd_yield Ibs/acr vs P rate for Different Irrigation Levels', fontsize=15)
plt.xlabel('P rate (Ibs/acre)', fontsize=15)
plt.ylabel('Yield (Ibs/acre)', fontsize=15)
plt.xticks(rotation=0, fontsize=15)
plt.yticks(fontsize=15)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['left'].set_color('black')
plt.legend(title='Irrigation Level', fontsize=12)
plt.ylim(2000, 10000)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
import scipy.stats

# Load your Excel data into a DataFrame
# Replace 'your_excel_file.xlsx' with the actual path to your Excel file
data = pd.read_excel('Growers 2023 24.xlsx', sheet_name='1st and 2nd')

# Calculate the mean and standard error for each Prate and IRR combination
mean_data = data.groupby(['Prate', 'IRR'], as_index=False).agg({'3rd_yield Ibs/acr': ['mean', 'sem']})
mean_data.columns = ['Prate', 'IRR', '3rd_yield Ibs/acr', '3rd_yield Ibs/acr_SE']

# Define the colors for the bars
colors = ['#1f77b4', '#ff7f0e']  # Colors for the different irrigation levels

# Create a scatter plot with error bars
plt.figure(figsize=(10, 8))

# Scatter plot for mean 3rd_yield Ibs/acr values with error bars
for irr, color in zip(mean_data['IRR'].unique(), colors):
    subset = mean_data[mean_data['IRR'] == irr]
    plt.errorbar(subset['Prate'], subset['3rd_yield Ibs/acr'], yerr=subset['3rd_yield Ibs/acr_SE'], fmt='o',
                 label=f'{irr} Irrigation', capsize=5, color=color)

    # Function to fit a piecewise linear model
    def piecewise_linear(x, a1, b1, a2, b2):
        return np.piecewise(x, [x < 50, x >= 50], [lambda x: a1 * x + b1, lambda x: a2 * x + b2])

    # Fit piecewise linear model
    X = subset['Prate'].values
    y = subset['3rd_yield Ibs/acr'].values
    params, pcov = curve_fit(piecewise_linear, X, y)

    # Calculate R-squared
    residuals = y - piecewise_linear(X, *params)
    ss_res = np.sum(residuals ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    r2 = 1 - (ss_res / ss_tot)

    # Plot the piecewise linear fit
    x_line = np.linspace(min(X), max(X), 100)
    plt.plot(x_line, piecewise_linear(x_line, *params), label=f'Piecewise Linear Fit (R²={r2:.2f})', color=color)

plt.title('Mean 3rd_yield Ibs/acr vs P rate for Different Irrigation Levels', fontsize=15)
plt.xlabel('P rate (Ibs/acre)', fontsize=15)
plt.ylabel('Yield (Ibs/acre)', fontsize=15)
plt.xticks(rotation=0, fontsize=15)
plt.yticks(fontsize=15)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['left'].set_color('black')
plt.legend(title='Irrigation Level', fontsize=12)
plt.ylim(2000, 10000)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load your Excel data into a DataFrame
# Replace 'your_excel_file.xlsx' with the actual path to your Excel file
data = pd.read_excel('Growers 2023 24.xlsx', sheet_name='1st and 2nd')

# Calculate the mean for each Prate and IRR combination
mean_data = data.groupby(['Prate', 'IRR'], as_index=False)['3rd_yield Ibs/acr'].mean()

# Define the colors for the bars
colors = ['#1f77b4', '#ff7f0e']  # Colors for the different irrigation levels

# Create a scatter plot
plt.figure(figsize=(10, 8))

# Scatter plot for mean 3rd_yield Ibs/acr values
for irr, color in zip(mean_data['IRR'].unique(), colors):
    subset = mean_data[mean_data['IRR'] == irr]
    plt.scatter(subset['Prate'], subset['3rd_yield Ibs/acr'], label=f'{irr} Irrigation', color=color)

    # Fit linear regression model
    X = subset['Prate'].values.reshape(-1, 1)
    y = subset['3rd_yield Ibs/acr'].values
    model = LinearRegression().fit(X, y)

    # Calculate R-squared
    r2 = model.score(X, y)

    # Plot the linear fit
    plt.plot(X, model.predict(X), label=f'Linear Fit (R²={r2:.2f})', color=color)

plt.title('Mean 3rd_yield Ibs/acr vs P rate for Different Irrigation Levels', fontsize=15)
plt.xlabel('P rate (Ibs/acre)', fontsize=15)
plt.ylabel('Yield (Ibs/acre)', fontsize=15)
plt.xticks(rotation=0, fontsize=15)
plt.yticks(fontsize=15)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['left'].set_color('black')
plt.legend(title='Irrigation Level', fontsize=12)
plt.ylim(2000, 10000)
plt.tight_layout()
plt.show()
```
